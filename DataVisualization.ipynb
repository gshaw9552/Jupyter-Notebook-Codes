```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars_data = pd.read_csv("toyota.csv", index_col=0, na_values=["??", "????"])
cars_data.shape
```

```python
cars_data.dropna(axis = 0, inplace=True)
cars_data.shape
```

```python
plt.scatter(cars_data['Age'], cars_data['Price'], c='red')
plt.title('Scatter Plot of Price vs Age of the cars')
plt.xlabel('Age (months)')
plt.ylabel('Price (Euros)')
plt.show()
```

```python
plt.hist(cars_data['KM'], color = 'green', edgecolor = 'white', bins = 5)
plt.title("Histogram of Kilometer")
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
fuel_counts = cars_data['FuelType'].value_counts()
fuel_list = [fuel_counts.get('Petrol', 0), fuel_counts.get('Diesel', 0), fuel_counts.get('CNG', 0)]
fuelType = ('Petrol', 'Diesel', 'CNG')
index = np.arange(len(fuelType))
plt.bar(index, fuel_list, color=['red', 'blue', 'cyan'])
plt.title("Bar plot of Fuel Types")
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index, fuelType, rotation = 90)
plt.show()
```
