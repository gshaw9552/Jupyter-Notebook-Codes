```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars_df = pd.read_csv('mtcars.csv')
```

```python
plt.scatter(cars_df['wt'], cars_df['mpg'])
plt.xlabel('wt')
plt.ylabel('mpg')
plt.show()
```

```python
diamond_df = pd.read_csv('diamond.csv')
```

```python
diamond_df.boxplot(column='price', by='cut')
```

```python
median_prices = diamond_df.groupby('cut')['price'].median()

max_median_cut = median_prices.idxmax()
max_median_price = median_prices.max()

print(f"The category with the highest median price is '{max_median_cut}' with a median price of {max_median_price}.")
```
